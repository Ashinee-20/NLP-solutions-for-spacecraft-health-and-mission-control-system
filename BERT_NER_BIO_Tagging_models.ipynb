```
import pandas as pd
data = pd.read_json("/content/drive/MyDrive/Datasets/NLP PROJECT/data.json")
```

```
data.head(30)
```

```
data['action'] = list(x['ACTION'] for x in list(data['BIO_tags']))
data['table'] = list(x['TABLE'] for x in list(data['BIO_tags']))
data.head()
```

```
data['tokens'][0]
```

```
tokens = []
actions = []
tables = []
sent = []
j = 0
for i in data['tokens']:
    tokens += i
    sent += [j]*(len(i))
    j += 1

for i in data['action']:
    actions += i

for i in data['table']:
    tables += i

print(len(sent),len(tokens), len(actions), len(tables))
```

```
a = data['tokens']
b = data['table']
c = data['action']

for i in range(len(a)):
  if len(a[i]) != len(b[i]) or len(a[i]) != len(b[i]):
      print(i, a[i])
```

```
df = pd.DataFrame()
df['sent'] = sent
df['tok'] = tokens
df['act'] = [x.upper() for x in actions]
df['tab'] = [x.upper() for x in tables]
df.head()
```

```
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```
X= df[["sent","tok"]]
Y =df["act"]
```

```
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size =0.2)
```

```
#building up train data and test data
train_data = pd.DataFrame({"sentence_id":x_train["sent"],"words":x_train["tok"],"labels":y_train})
test_data = pd.DataFrame({"sentence_id":x_test["sent"],"words":x_test["tok"],"labels":y_test})
```

```
train_data.head()
```

### Actions and Features NER Extraction

```
from simpletransformers.ner import NERModel,NERArgs
```

```
label_act = df["act"].unique().tolist()
label_act
```

```
args = NERArgs()
args.num_train_epochs = 20
args.learning_rate = 1e-4
args.overwrite_output_dir =True
args.train_batch_size = 32
args.eval_batch_size = 32
```

```
model_act = NERModel('bert', 'bert-base-cased',labels=label_act,args =args)
```

```
model_act.train_model(train_data,eval_data = test_data,acc=accuracy_score)
```

```
result, model_outputs, preds_list = model_act.eval_model(test_data)
```

```
result
```

```
import joblib

joblib.dump(model_act, 'features_model.pkl')
```

```
model_features = joblib.load('/content/drive/MyDrive/Datasets/NLP PROJECT/features_model.pkl')

prediction, model_output = model_features.predict(["show me bar graph of temperATURE V/S pressure"])
```

```
label_act.remove('O')
```

```
features = []

for i in prediction[0]:
    for j in i:
        if(i[j] in label_act and i[j] != 'B-ACTION' and i[j] != 'I-ACTION'):
                features.append(j)


print(features)
```
