```python
import sqlite3, random
import pandas as pd
from datetime import datetime, timedelta

conn = sqlite3.connect('alpha.db')
cursor = conn.cursor()

schema = '''
    CREATE TABLE IF NOT EXISTS alpha_table (
        timestamp DATE,
        temperature REAL,
        pressure REAL,
        altitude REAL,
        power REAL,
        voltage REAL,
        fuellevel REAL,
        latitude REAL,
        longitude REAL,
        thrust REAL,
        commstatus TEXT,
        enginetemp REAL,
        battvoltage REAL,
        solarpower REAL,
        fuelflow REAL,
        oxygenlevel REAL,
        gyrox REAL,
        gyroy REAL,
        gyroz REAL,
        statusflags TEXT
    )
'''

cursor.execute(schema)

data = []
r=1000
current_time = datetime.now()
ref_time = (current_time - timedelta(minutes=30))

for i in range(r):
    timestamp = (ref_time + timedelta(minutes=i)).strftime('%Y-%m-%d %H:%M:%S')
    row_data = [timestamp] + [random.uniform(10.0, 100.0) for _ in range(19)]
    data.append(tuple(row_data))


cursor.executemany('INSERT INTO alpha_table VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)', data)
conn.commit()
conn.close()

conn = sqlite3.connect('alpha.db')
df = pd.read_sql_query('SELECT * FROM alpha_table', conn)
conn.close()

df
```
